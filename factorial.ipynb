```python
from time import time
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from sympy import *
import numpy as np
```

```python
def factorial_it(n):
    produ = 1
    while n>0:
        produ = produ * n 
        n = n-1
    return produ
```

```python
def factorial_recur(n):
    if n <= 1:
        return 1
    else:
        return n * factorial_recur(n-1)
```

```python
tabla = pd.DataFrame()
tabla["Valor"] = None
tabla["iterative"] = None
tabla["recursive"] = None

valor = []
iterative = []
recursive = []
matrixTime = []
for i in range(0, 1000):
    valor.append(i)

    inicio = time()
    factorial_it(i)
    final = time()
    iterative.append(final - inicio)


    inicio = time()
    factorial_recur(i)
    final = time()
    recursive.append(final - inicio)


tabla["Valor"] = valor
tabla["iterative"] = iterative
tabla["recursive"] = recursive
tabla
```

```python
plt.style.use("bmh") 
eje_xiter = tabla["iterative"]
eje_xrecur = tabla['recursive']
eje_y = tabla['Valor']
plt.title("RecursiÃ³n")
plt.plot(eje_y, eje_xiter, color = "red", label='linearFib') #Iterable
plt.plot(eje_y, eje_xrecur, color = "green", label='matrices') #RecursiÃ³n

plt.xlabel('Valor')
plt.ylabel('Tiempo')

plt.show()
```
