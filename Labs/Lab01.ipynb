```python
import random
from time import time
import pandas as pd
import matplotlib.pyplot as plt
from math import log
```

```python
tabla = pd.DataFrame()
tabla["longitud"] = None
tabla["tiempo"] = None
print(tabla)
```

```python
longitud = []
tiempo = []
for longitud_int in range(1,10000):
    l = []
    longitud.append(longitud_int)
    
    for i in range(0, longitud_int):
        l += [random.randint(1, 1000)]
        
    inicio = time()
    l.sort()
    final = time()
    
    duración = final - inicio
    tiempo.append(duración)
```

```python
tabla["longitud"] = longitud
tabla["tiempo"] = tiempo
```

```python
tabla = tabla.sort_values('longitud')
```

```python
tabla
```

```python
plt.style.use("bmh") 
eje_x = tabla['tiempo']
eje_y = tabla['longitud'] 
```

```python
plt.plot(eje_y, eje_x, color = "blue")
plt.xlabel('Longitud de la lista')
plt.ylabel('Tiempo')

plt.show
```

```python
def f1(x):
    return x * log(x)

def f2(x):
    return x
  
rango = range(1,10000)
plt.plot(rango, [f1(i) for i in rango], color = "red")
plt.plot(rango, [f2(i) for i in rango], color = "black")
plt.show
```
