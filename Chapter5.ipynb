```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import ctypes

class DynamicArray:

    def __init__(self):

        self.datos = 0
        self._capacity = 1
        self._A = self._make_array(self._capacity)

    def __len__(self):
        return self.datos 

    def __getitem__(self, k):    
        #if  not 0<=k<self.datos:
           # raise IndexError('invalid index') 
        return self._A[k]

    def append(self, obj):
        if self.datos == self._capacity:
            self._resize(2  * self._capacity)
        self._A[self.datos] = obj
        self.datos += 1

    def pop(self):
        if self.datos < (self._capacity)/4:
            self._resize(0.5 * self._capacity)
        self._A = self._A[0:(self.datos-1)]
        self.datos -= 1
            

    def  _resize(self, c):
        B = self._make_array(c)
        for k in range(self.datos):
            B[k] = self._A[k]
        self._A = B
        self._capacity = c

    def _make_array(self, c):
        return (c  * ctypes.py_object)( )




    
```

```python
# Ejercicio 5.3

agregar = pd.DataFrame()
agregar["Lenght"] = None
agregar["Size"] = None
import sys
n = 100
data = []
lenght = []
size = []
for k in range(n):
    a = len(data)
    lenght.append(a)
    b = sys.getsizeof(data)
    size.append(b)

    print( "Lenght: {0:3d};  Size in bytes: {1:4d}".format(a,b))

    data.append(None)

agregar["Lenght"] = lenght
agregar["Size"] = size
```

```python
quitar = pd.DataFrame()
quitar["Lenght"] = None
quitar["Size"] = None

lenght_q = []
size_q = []
import sys
n = 100
data = [None for k in range(n)]
for k in range(n):
    data.pop(-1)
    a = len(data)
    lenght_q.append(a)
    b = sys.getsizeof(data)
    size_q.append(b)

    print( "Lenght: {0:3d};  Size in bytes: {1:4d}".format(a,b))



quitar["Lenght"] = lenght_q
quitar["Size"] = size_q
```

```python
plt.style.use("bmh") 


ejex = agregar["Lenght"]
ejey = agregar["Size"]

ejex_q = quitar["Lenght"]
ejey_q = quitar["Size"]

fig, ax = plt.subplots()
ax.step(ejex, ejey, color = "blue", )
ax.step(ejex_q, ejey_q, color = "#990000")

plt.title("Tamaño en memoria por cantidad de datos")
plt.xlabel("Largo")
plt.ylabel("Tamaño en memoria")
```
